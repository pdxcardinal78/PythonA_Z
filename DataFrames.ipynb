```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
#Distribution
vis1 = sns.distplot(stats.InternetUsers)
```

```python
vis1 = sns.distplot(stats.InternetUsers, bins =30)
```

```python
#Boxplot
vis2 = sns.boxplot(data = stats, x='IncomeGroup', y='BirthRate')
```

```python
vis3 = sns.lmplot('InternetUsers', 'BirthRate', stats, fit_reg = False)
```

```python
vis3 = sns.lmplot('InternetUsers', 'BirthRate', stats, fit_reg = False, \
                 hue='IncomeGroup', size = 8,)
```
