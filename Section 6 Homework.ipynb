```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv('Section6-Homework-Data.csv', encoding='latin-1')
```

```python
movies.describe()
```

```python
movies.info()
```

```python
movies.Genre = movies.Genre.astype('category')
movies.Studio = movies.Studio.astype('category')
```

```python
movies.info()
```

```python
g = movies.groupby('Genre')
g1 = g.count()
g1
```

```python
g = movies.groupby('Studio')
g.count()
```

```python
studios = ['WB', 'Universal', 'Sony', 'Paramount Pictures', 'Fox', 'Buena Vista Studios']
genres = ['action', 'adventure', 'animation', 'comedy', 'drama']
movies1 = movies[movies['Studio'].isin(studios)]
movies2 = movies1[movies1['Genre'].isin(genres)]
movies3 = movies2[['Studio', 'Genre', 'Gross % US']]
movies3.Genre.cat.remove_unused_categories(inplace=True)
movies3.Studio.cat.remove_unused_categories(inplace=True)
```

```python
movies3
```

```python
f, ax = plt.subplots(figsize=(12,8))
sns.set_style('whitegrid')

# Draw a categorical scatterplot to show each observation
sns.swarmplot(x='Genre', y='Gross % US', hue='Studio', data=movies3)
sns.boxplot(x='Genre', y='Gross % US', data=movies3, color='lightgrey')
plt.legend(frameon=True, shadow=True, prop={'size':10}, loc=1, bbox_to_anchor=(1.25, 1))
plt.xlabel('Genre', fontsize=30)
plt.ylabel('Gross % US', fontsize=30)
plt.title('Gross % Sales by Genre', fontsize=40)
plt.show()
```
